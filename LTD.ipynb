```python
#  Defining the function that will contain the parameters of the names and all the courses
def bts_choices(name, courses):
  # Creating a course dictionary that holds all of the courses
  course_dict = dict()
  # Creating a list of all the course details
  course_deets=[]
  nums=0
  # Assigning numerical values to the courses
  for num,course in enumerate(courses):
    course_dict.setdefault(num+1,course)
    # Printing the value and the course name
    print(str(num+1)+" "+course["Course_Name"])
    # Checking to see the user input
    nums=num
  i = int(input("Which course: "))
  # Getting the chosen course of the user
  if (i<=nums):
    course_deets=list(course_dict.get(i))
  else:
    bts_choices(name,courses)
    # Asking which detail the user wants to know

  choice=1
  while(choice!=0):
    for num,course_deet in enumerate(course_deets):
      nums=num
      print(str(num+1)+" "+course_deet)
    print("0 Exit\n")
    choice = int(input("Please choose next {} : ".format(name)))
    # Asking the user to choose between the range
    if(choice>nums+1):
      print("Please choose between 1 and {}".format(nums))
    print("\n")
    # If the user chooses from the range, getting his preffered course details
    if(choice<=nums and choice!=0):
      print(course_deets[choice-1]+" : "+course_dict.get(i).get(course_deets[choice-1])+"\n\n")
      # Asking the user if he wants to register
  reg = input("Do you want to register(y/n): ")
  # If user registers, increase the count by 1
  if(reg == 'y' or reg == "yes"):
    global count
    count = count+1
    # If user says no, ask if he wants to browse the course again or else exit
  if(reg=='n' or reg=="no"):
    again = input("do you browse the courses again(y/n): ")
    if(again=='y' or reg=="yes"):
      return choices(name)
    if(again=='n' or again=="no"):
      return
  
```
