Wordcloud for Yelp reviews

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
yelp = pd.read_csv(io.BytesIO(uploaded['yelp.csv']))
```

```python
yelp.head()
```

```python
# Dropping unwanted details
yelp = yelp["text"]
```

```python
yelp.head()
```

```python
# Data preprocessing
review_words = ''
stopwords = set(STOPWORDS)
```

```python
for val in yelp.values:
  val = str(val)
  tokens = val.split(" ")
  for token in tokens:
    token = token.lower()
  review_words += " ".join(tokens)+" "
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                      background_color = 'white',
                      stopwords = stopwords,
                      min_font_size = 10).generate(review_words)
```

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
```
